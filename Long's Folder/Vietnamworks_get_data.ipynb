```python
options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(options=options)
```

```python
from selenium import webdriver
import time

driver.get("https://www.vietnamworks.com/viec-lam?g=5")

# Initialize the WebDriver
# # # Function to scroll and load content on the current page
def scroll_to_load():
    SCROLL_PAUSE_TIME = 2
    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(SCROLL_PAUSE_TIME)
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

    soup = BeautifulSoup(driver.page_source, 'html.parser')
    # print(soup.prettify())

def extract_data():
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    divs = soup.find_all('div', class_='sc-nZgfj jSXiyi')

    # Extract links from the <a> tags within these divs
    job_links = []
    for div in divs:
        a_tag = div.find('a', href=True)
        if a_tag:
            job_links.append(a_tag['href'])

    unique_links = set(job_links)
    return unique_links

all_links = set()  # To store all unique links across all pages

while True:
    try:
        # Scroll to load all content on the current page
        scroll_to_load()

        links = extract_data()
        all_links.update(links)

        # time.sleep(2)

        # Locate the "Next" button
        next_button_xpath = "//button[text()='>']"
        # Wait for the button to be visible (not just present in the DOM)
        next_button = WebDriverWait(driver, 30).until(
            EC.element_to_be_clickable((By.XPATH, next_button_xpath))
        )

        print(next_button.get_attribute("outerHTML"))
        # Check if the button is disabled or no longer present
        if not next_button.is_enabled():
            print("Reached the last page.")
            break

        # Scroll the "Next" button into view
        driver.execute_script("arguments[0].scrollIntoView();", next_button)
        next_button.click()

        # Wait for the next page to load
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.TAG_NAME, 'body'))
        )
    except TimeoutException as e:
        print(f"Timeout error: {e}")
        break
    except NoSuchElementException as e:
        print(f"Element not found error: {e}")
        break
    except Exception as e:
        print(f"An error occurred: {e}")

# Output all unique links
print(f"Total unique links collected: {len(all_links)}")
for link in all_links:
    print(link)
```

```python
def scrape_link(link):
    driver.get(link)
    time.sleep(3) # wait for page to load - just to be safe
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    try:
        xpath = '//div[h2[text()="Thông tin việc làm"]]//button[text()="Xem thêm"]' #Need to find this specific button
        WebDriverWait(driver, 10).until(
          EC.element_to_be_clickable((By.XPATH, xpath))
        )
        button = driver.find_element(By.XPATH, xpath)
        driver.execute_script("arguments[0].scrollIntoView();", button)
        button.click()
    except Exception as e:
        print(f"Error clicking the button: {e}")
        return None

    time.sleep(5)  # Wait for new content to load after the click
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    skills = None
    role = None
    level = None
    year_of_experiences = None

    #TITLE
    div = soup.find('div', class_='sc-8868b866-0 dvidDw')
    job_title = div.find('h1', class_='sc-df6f4dcb-0 bsKseP')

    #SALARY
    salary = soup.find('span', class_='sc-df6f4dcb-0 iOaLcj')

    div = soup.find('div', class_="sc-37577279-4 kNdlhJ")
    #LOCATION
    location = div.find('p') if div else None #the location appear first so we don't have to worry about duplicate or wrong

    #COMPANY NAME
    company = soup.find('a', class_="sc-df6f4dcb-0 dIdfPh sc-f0821106-0 gWSkfE")

    #LEVEL
    pattern = re.compile(r'\b(?:' + '|'.join(LEVEL_KEYWORDS) + r')\b', re.IGNORECASE)
    level = pattern.findall(job_title.get_text(strip=True))

    #ROLE
    # Regular expression to match words outside of (), [], {}, and similar brackets
    pattern = re.compile(r'[^()\[\]{}<>]+')
    # Find all matches
    words = pattern.findall(job_title.get_text(strip=True))
    # Clean up the result by stripping any leading/trailing whitespace
    words = [word.strip() for word in words if word.strip()]
    role = ' '.join(words)

    #SKILL AND OTHER STUFF
    divs = soup.find_all('div', class_='sc-8868b866-0 GVIEn')
    for div in divs:
      labels = div.find_all('label', class_="sc-df6f4dcb-0 fgNUpS")  # Find the label within each div
      for label in labels:
          if label and 'KỸ NĂNG' in label.get_text():
              skills = div.find('p')  # Find the <p> tag with the skills
          if label and 'SỐ NĂM KINH NGHIỆM TỐI THIỂU' in label.get_text():
              year_of_experiences = div.find('p')  # Find the <p> tag with the skills

    source_platform = link

    return {
        'Job Title': job_title.get_text(strip=True) if job_title else 'N/A',
        'Role': role if role else 'N/A',
        'Level': level if level else 'N/A',
        'Year Of Experiences': year_of_experiences.get_text(strip=True) + " năm kinh nghiệm" if year_of_experiences else 'N/A',
        'Company': company.get_text(strip=True) if company else 'N/A',
        'Location': location.get_text(strip=True) if location else 'N/A',
        'Salary Range': salary.get_text(strip=True) if salary else 'N/A',
        'Required Skills': skills.get_text(strip=True) if skills else 'N/A',
        'Source Platform': source_platform
    }
```

```python
#Gather data
data = []
for link in links:
    try:
        print(f"Scraping: {link}")
        result = scrape_link(link)
        if(result != None):
          data.append(result)
        time.sleep(2)
    except Exception as e:
        print(f"Error scraping {link}: {e}")
```

```python
import pandas as pd
import ast

def clean_level_column(file_path):
    # Read the CSV file
    df = pd.read_csv(file_path)

    def clean_level(value):
        # Handle None and empty values
        if pd.isna(value) or value == "" or str(value).strip() == "":
            return "N/A"

        # Convert to string if not already
        value_str = str(value).strip()

        try:
            # First attempt: Try to evaluate as literal list/array
            cleaned = ast.literal_eval(value_str)
            if isinstance(cleaned, list):
                return cleaned[0] if cleaned and cleaned[0] else N/A
        except (ValueError, SyntaxError):
            pass

        # Remove brackets, quotes, and extra whitespace
        cleaned = re.sub(r"[\[\]'\"]", "", value_str).strip()

        # Remove any remaining special characters and extra spaces
        cleaned = re.sub(r'\s+', ' ', cleaned)

        return cleaned if cleaned else N/A

    # Apply the cleaning function to the Level column
    df['Level'] = df['Level'].apply(clean_level)

    return df

# Example usage
if __name__ == "__main__":
    file_path = '/content/job_data.csv'

    # Clean the data
    cleaned_df = clean_level_column(file_path)

    # Save the cleaned data to a new CSV file
    cleaned_df.to_csv('cleaned_job_data.csv', index=False)
    print("Data cleaning completed. Check 'cleaned_job_data.csv' for results.")
```
